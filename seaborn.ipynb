```python
import pandas as pd
import pydataset
```

```python
pydataset.data('titanic', show_doc=True)
```

```python
titanic = pydataset.data('titanic')
```

```python
titanic.head()
```

```python
titanic['class'].value_counts().plot(kind = 'bar')
```

```python
titanic['survived'].value_counts().plot(kind = 'bar')
```

```python
titanic.groupby('survived')['class'].value_counts().plot(kind = 'bar')
```

```python
states = pd.read_csv('populacao_brasil.csv')
```

```python
states.head()
```

```python
states.total.hist()
```

```python
fig, ax = plt.subplots()
plt.hist(states.total, bins = 15, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
states['percentual'] = states.total / states.total.sum()
```

```python
states.head()
```

```python
plt.pie(states.percentual, labels=states.estado, autopct='%1.2f%%')
plt.show()
```
