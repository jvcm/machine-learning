```python
iris = pydataset.data('iris')
```

```python
def specie_color(x):
    if x == 'setosa':
        return 0
    if x == 'versicolor':
        return 1
    return 2

iris['SpeciesNumber'] = iris['Species'].apply(specie_color)

iris.head()
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'],
            c = iris.SpeciesNumber,
            cmap = 'vlag',
            alpha = 0.4)
```
