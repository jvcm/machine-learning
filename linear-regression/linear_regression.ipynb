```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import pydataset
import matplotlib.pyplot as plt

sns.set()
```

```python
class LinearRegression:
    def __init__(self, data):
        self.data = data
        self.att_Parameters(data)
        return
    
    def att_Parameters(self, data):
        X = np.matrix(data.drop(axis = 1, columns = ['Value']))
        Y = np.matrix(data.Value).T
        A = np.matmul(X.T, X)
        A_inv = np.linalg.pinv(A)
        B = np.matmul(X.T, Y)
        param = np.matmul(A_inv, B)
        self.parameters = param

    def get_Parameters(self):
        return self.parameters
```

```python
df = pd.read_csv('yacht-hydrodynamics.csv', sep=' ')
df['Value'] = df['resid-resist']
del df['resid-resist']
df.head()
```

```python
ln_model = LinearRegression(df)
ln_model.get_Parameters()
```

```python
x = np.array(df.drop(axis = 1, columns = ['Value']))
y = np.array(df['Value'])
b = ln_model.get_Parameters()

for i, v in enumerate(x):
    print('y =',y[i],' - f(x) =', np.dot(b.T, v))
```
